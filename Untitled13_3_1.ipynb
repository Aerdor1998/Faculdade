#aprendizado não supervisionado


```python
# aprendizado não-supervisionada cluster
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

dados = pd.read_csv('produtos_agricolas.csv')

colunas = ['N', 'P', 'K', 'temperature', 'humidity', 'ph', 'rainfall']
dados_selecionados = dados[colunas]

scaler = StandardScaler()
dados_padronizados = scaler.fit_transform(dados_selecionados)

kmeans = KMeans(n_clusters=3, random_state=0)
clusters = kmeans.fit_predict(dados_padronizados)

dados['Cluster'] = clusters

plt.scatter(dados['temperature'], dados['humidity'], c=dados['Cluster'], cmap='viridis')
plt.xlabel('Temperatura')
plt.ylabel('Umidade')
plt.title('Análise de Clusters - Preferências Climáticas')
plt.show()

```

#modelos preditivos

```python
import pandas as pd
dados = pd.read_csv("produtos_agricolas.csv")

X_proda = dados.drop("label", axis=1)  
y_proda = dados["label"]  
```

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

modelo_reglog = LogisticRegression()
modelo_reglog.fit(X_proda, y_proda)

previsoes_reglog = modelo_reglog.predict(X_proda)
acuracia_reglog = accuracy_score(y_proda, previsoes_reglog)
```

```python
#arvore de decisão
from sklearn.tree import DecisionTreeClassifier

modelo_arvore = DecisionTreeClassifier()
modelo_arvore.fit(X_proda, y_proda)

previsoes_arvore = modelo_arvore.predict(X_proda)
acuracia_arvore = accuracy_score(y_proda, previsoes_arvore)
```

```python
#random forest
from sklearn.ensemble import RandomForestClassifier

modelo_rf = RandomForestClassifier()
modelo_rf.fit(X_proda, y_proda)

previsoes_rf = modelo_rf.predict(X_proda)

acuracia_rf = accuracy_score(y_proda, previsoes_rf)
```

```python
from sklearn.neural_network import MLPClassifier

modelo_mlp = MLPClassifier(hidden_layer_sizes=(100, 100), max_iter=500)
modelo_mlp.fit(X_proda, y_proda)

previsoes_mlp = modelo_mlp.predict(X_proda)

acuracia_mlp = accuracy_score(y_proda, previsoes_mlp)
```

```python
from sklearn.neural_network import MLPClassifier

modelo_mlp_personalizado = MLPClassifier(hidden_layer_sizes=(200, 100, 50), max_iter=1000)
modelo_mlp_personalizado.fit(X_proda, y_proda)

previsoes_mlp_personalizado = modelo_mlp_personalizado.predict(X_proda)

acuracia_mlp_personalizado = accuracy_score(y_proda, previsoes_mlp_personalizado)
```
